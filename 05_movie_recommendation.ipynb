* 영화1: 평점이 10/10인데 5명만 평가
* 영화2: 평점이 8/10인데 500명이 평가  

이 두 영화 중 더 신뢰도가 높은 정보는 영화2의 정보일 겁니다. 따라서 여기서는 vote_average와 vote_count를 동시에 고려해야 합니다.  
![가중치.png](https://image.ibb.co/jYWZp9/wr.png)

````python
C = df2['vote_average'].mean() # 평점 평균
C
````

````python
m = df2['vote_count'].quantile(0.9) # 상위 10% 데이터 추출 - 이보다 더 많은 평가를 받은 데이터를 뽑을 예정.
m
````

````python
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
````

````python
q_movies['vote_count'].sort_values()
````

````python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (m + v) * C)
````

````python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies.head(3)
````

````python
q_movies = q_movies.sort_values('score', ascending=False)
````

````python
q_movies[['title', 'vote_average', 'vote_count', 'score']].head(10)
````

````python
import matplotlib.pyplot as plt
````

````python
pop= df2.sort_values('popularity', ascending=False)

plt.figure(figsize=(15,7))
plt.barh(pop['title'].head(10),pop['popularity'].head(10), align='center',
        color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel("Popularity")
plt.title("Popular Movies", fontsize=18)
````

## 2. Content Based Filtering (컨텐츠 기반 필터링)

````python
# 영화 줄거리 (정확히는 시놉시스)
df2['overview'].head(5)
````

Bag of Words(BoW) 기법 적용을 위해 문서 내의 모든 단어를 임베딩 - 벡터화하겠습니다.  
1. TfidVectorizer (TF-IDF 기반의 벡터화)

````python
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
tfidf = TfidfVectorizer(stop_words='english')
````

````python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
````

````python
ENGLISH_STOP_WORDS # 영어 불용어 목록
````

````python
df2['overview'] = df2['overview'].fillna('')
````

````python
tfidf_matrix = tfidf.fit_transform(df2['overview'])
tfidf_matrix.shape
````

````python
tfidf_matrix
````

### 단어 유사도 - 코사인 유사도

````python
# cosine_similarity 함수를 임포트해도 되지만, 공식 문서에 따르면 속도 측면에서 linear_kernel 함수를 사용하는 것이 더 유리하다고 합니다.
from sklearn.metrics.pairwise import linear_kernel
````

````python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
````

이 행렬에서 각 행과 열은 문장 1, 2, ..., $n$을 뜻하고, 행렬의 각 요소는 코사인 유사도를 뜻합니다.

````python
cosine_sim.shape
````

````python
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
indices
````

````python
indices['The Dark Knight Rises']
````

````python
df2.iloc[[3]]
````

````python
test_idx = indices['The Dark Knight Rises'] # 영화 제목을 통해 전체 데이터 기준 그 영화의 index 값을 얻기
test_idx
````

````python
cosine_sim[3]
````

````python
test_sim_scores = list(enumerate(cosine_sim[3])) # 코사인 유사도 행렬에서 idx에 해당하는 데이터를 (idx, 유사도) 형태로 얻어옴.
````

````python
test_sim_scores = sorted(test_sim_scores, key=lambda x: x[1], reverse=True) # 코사인 유사도 기준으로 내림차순 정렬
test_sim_scores[1:11] # 자기 자신을 제외한 10개의 추천 영화를 슬라이싱
````

````python
# 추천 영화 목록 10개의 인덱스 정보 추출
test_movie_indices = [i[0] for i in test_sim_scores[1:11]]
test_movie_indices
````

````python
# 인덱스 정보를 통해 영화 제목 추출
df2['title'].iloc[test_movie_indices]
````

자, 여기까지의 과정을 묶어서 하나의 함수로 정의하겠습니다.

````python
# 영화 제목을 입력받아 코사인 유사도를 통해 가장 유사도가 높은 상위 10개의 영화 목록 반환
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    
    movie_indices = [i[0] for i in sim_scores]
    
    return df2['title'].iloc[movie_indices]
````

````python
get_recommendations('Avengers: Age of Ultron')
````

````python
get_recommendations('The Avengers')
````

겉보기에는 딕셔너리 형태로 데이터가 주어진 것 같지만, `type` 내장 함수를 이용해보니 `str`, 즉 문자열 데이터라는 진단이 나왔습니다. 따라서 우리는 이 문자열 데이터를 실제 딕셔너리 데이터로 만드는 작업을 거쳐야 합니다. 여기서 우리는 `ast` 패키지의 `literal_eval` 함수를 쓸 수 있습니다.

````python
s1 = [{"id": 28, "name": "Action"}]
s2 = '[{"id": 28, "name": "Action"}]'
````

````python
type(s1), type(s2)
````

````python
from ast import literal_eval
````

````python
s2 = literal_eval(s2)
s2, type(s2)
````

````python
print(s1)
print(s2)
````

````python
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
````

````python
df2.loc[0, 'crew']
````

````python
# 감독 정보를 추출
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
````

````python
df2['director'] = df2['crew'].apply(get_director)
df2['director']
````

````python
df2[df2['director'].isnull()]
````

````python
df2.loc[0, 'cast']
````

````python
df2.loc[0, 'genres']
````

````python
# 처음 3개의 데이터 중 name에 해당하는 value만 추출
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
````

````python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
````

````python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
````

````python
# 소문자로 변환 및 띄어쓰기 제거
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    elif isinstance(x, str):
            return str.lower(x.replace(' ', ''))
    else:
        return ''
````

````python
features = ['cast', 'keywords','director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
````

````python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
````

````python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
````

````python
df2['soup'] = df2.apply(create_soup, axis=1)
````

````python
df2['soup']
````

````python
df2['overview']
````

````python
from sklearn.feature_extraction.text import CountVectorizer
````

````python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
````

````python
count_matrix
````

````python
from sklearn.metrics.pairwise import cosine_similarity
````

````python
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
cosine_sim2
````
